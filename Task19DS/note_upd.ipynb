```python
import numpy as np
import pandas as pd
```

```python
attend = pd.read_csv('datasets/attend.csv')
combine_df = pd.read_csv('datasets/combine_df.csv')
```

```python
combine_df
```

```python
new_attend = attend.drop(['уникальный номер участника', 'уникальный номер занятия'], axis=1).drop_duplicates(subset=['уникальный номер группы', 'направление 2', 'направление 3'], keep=False)
new_attend
```

```python
new_attend['уникальный номер занятия'].unique().shape
```

```python
new_attend['уникальный номер группы'].unique().shape
```

```python
omg_df = pd.merge(combine_df, new_attend, how='left', left_on=['id'], right_on=['уникальный номер группы'])
omg_df
```

```python
omg_df.drop(['уникальный номер группы', 'направление 2', 'направление 3', 'дата занятия', 'время начала занятия', 'время окончания занятия'], axis=1, inplace=True)
```

```python
omg_df
```

```python
omg_df.rename(columns={'онлайн/офлайн':'online'}, inplace=True)
```

```python
omg_df.to_csv('datasets/combine_upd_df.csv', encoding='utf-8', index=False)
```
